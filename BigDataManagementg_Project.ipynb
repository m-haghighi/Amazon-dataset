```python
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)

root_dir = "/content/gdrive/My Drive/amazon"
```

```python
import dask
import dask.dataframe as dd
```

```python
dask_df_VideoGames = dd.read_csv(root_dir+'/Video_Games.csv')
dask_df_VideoGames.columns=['Item','User','Rating','Timestamp']
dask_df_VideoGames['VideoGames']=1
dask_df_VideoGames=dask_df_VideoGames.drop(columns=['Item','Rating','Timestamp'])
dask_df_VideoGames=dask_df_VideoGames.drop_duplicates(subset = ["User"])
dask_df_VideoGames.head()
```

```python
dask_df_ToysandGames = dd.read_csv(root_dir+'/Toys_and_Games.csv')
dask_df_ToysandGames.columns=['Item','User','Rating','Timestamp']
dask_df_ToysandGames['ToysandGames']=1
dask_df_ToysandGames=dask_df_ToysandGames.drop(columns=['Item','Rating','Timestamp'])
dask_df_ToysandGames=dask_df_ToysandGames.drop_duplicates(subset = ["User"])
dask_df_ToysandGames.head()
```

```python
dask_df_ToolsandHomeImprovement = dd.read_csv(root_dir+'/Tools_and_Home_Improvement.csv')
dask_df_ToolsandHomeImprovement.columns=['Item','User','Rating','Timestamp']
dask_df_ToolsandHomeImprovement['ToolsandHomeImprovement']=1
dask_df_ToolsandHomeImprovement=dask_df_ToolsandHomeImprovement.drop(columns=['Item','Rating','Timestamp'])
dask_df_ToolsandHomeImprovement=dask_df_ToolsandHomeImprovement.drop_duplicates(subset = ["User"])
dask_df_ToolsandHomeImprovement.head()
```

```python
dask_df_Digital_Music = dd.read_csv(root_dir+'/Digital_Music.csv')
dask_df_Digital_Music.columns=['Item','User','Rating','Timestamp']
dask_df_Digital_Music['Digital_Music']=1
dask_df_Digital_Music=dask_df_Digital_Music.drop(columns=['Item','Rating','Timestamp'])
dask_df_Digital_Music=dask_df_Digital_Music.drop_duplicates(subset = (["User"]))
dask_df_Digital_Music=dask_df_Digital_Music.astype('object')
dask_df_Digital_Music.head()
```

```python
dask_df_Software = dd.read_csv(root_dir+'/Software.csv')
dask_df_Software.columns=['Item','User','Rating','Timestamp']
dask_df_Software['Software']=1
dask_df_Software=dask_df_Software.drop(columns=['Item','Rating','Timestamp'])
dask_df_Software=dask_df_Software.drop_duplicates(subset = (["User"]))
#dask_df_Software=dask_df_Software.astype('object')
dask_df_Software.head()
```

```python
dask_df_all= dd.merge(dask_df_ToysandGames, dask_df_VideoGames, how="outer", on=["User"])
dask_df_all= dd.merge(dask_df_all, dask_df_ToolsandHomeImprovement, how="outer", on=["User"])
dask_df_all= dd.merge(dask_df_all, dask_df_Digital_Music, how="outer", on=["User"])
dask_df_all= dd.merge(dask_df_all, dask_df_Software, how="outer", on=["User"])
dask_df_all=dask_df_all.fillna(0)
```

```python
dask_df_all.head(20)
```

```python
len(dask_df_all)
```

```python
%pip install mlxtend --upgrade
```

```python
#FP growth algorithm
#Importing Libraries
from mlxtend.frequent_patterns import fpgrowth

#running the fpgrowth algorithm
dask_df_all_WUser=dask_df_all.drop(columns=["User"])
res=fpgrowth(dask_df_all_WUser.compute(),min_support=0.01, use_colnames=True)

# printing top 10
res.head(10)
```

```python
#Association rules
# importing required module
from mlxtend.frequent_patterns import association_rules

# creating asssociation rules
res=association_rules(res, metric="lift", min_threshold=0.8)

# printing association rules
res
```

```python
res.sort_values("confidence",ascending=False)
```
